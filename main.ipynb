```python
import os

def reformat_character_positions(input_file_path, output_file_path):
    try:
        with open(input_file_path, 'r', encoding='utf-8') as input_file:
            lines = input_file.readlines()
    except UnicodeDecodeError as e:
        print(f"Error reading the input file: {e}")
        return

    formatted_lines = []
    current_question = None
    current_positions = []

    for line in lines:
        line = line.strip()

        # Check if the line starts with "Question Number"
        if line.startswith("Question Number"):
            # Write the previous question and its positions without adding extra blank lines
            if current_question and current_positions:
                formatted_lines.append(f"{current_question}")
                formatted_lines.append(',\n'.join(current_positions) + ',')
            
            # Reset for the new question
            current_question = line
            current_positions = []
        
        # Check if the line starts with "Character position"
        elif line.startswith("Character position"):
            parts = line.split(': ')
            if len(parts) == 2:
                position = parts[0].split()[2]  # Get the position index
                character = parts[1].strip().strip(",")  # Get the character
                current_positions.append(f"{position}: {character}")
    
    # Write the last question and its positions
    if current_question and current_positions:
        formatted_lines.append(f"{current_question}")
        formatted_lines.append(',\n'.join(current_positions) + ',')

    # Write the formatted output to the new file
    with open(output_file_path, 'w', encoding='utf-8') as output_file:
        for formatted_line in formatted_lines:
            output_file.write(formatted_line + '\n')

# Define the input and output folder paths
input_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\4_Clean Bullets"
output_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\5_Formatted Bullets"

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all text files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".txt"):
        input_file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, filename)

        # Call the function to reformat each file
        reformat_character_positions(input_file_path, output_file_path)

print("Reformatted text files have been saved to the '5_Formatted Bullets' folder.")
```

```python
import re
import os

# Function to extract above and below positions based on "Question Number"
def extract_above_below_positions(input_file_path):
    try:
        # Open and read the input file
        with open(input_file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print(f"Error: The file {input_file_path} was not found.")
        return None, None
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
        return None, None

    extracted_positions = ['0']  # Initialize with 0 as the starting position
    previous_position = None

    # Loop through each line in the file
    for i, line in enumerate(lines):
        line = line.strip()
        
        # Check if the line contains "Question Number"
        if line.startswith("Question Number"):
            # Find the position just above
            if i > 0:
                prev_line = lines[i - 1].strip()  # Previous line
                prev_match = re.match(r"(\d+):", prev_line)  # Match the position
                if prev_match:
                    previous_position = prev_match.group(1)

            # Find the position just below
            if i + 1 < len(lines):
                next_line = lines[i + 1].strip()  # Next line
                next_match = re.match(r"(\d+):", next_line)  # Match the position
                if next_match:
                    next_position = next_match.group(1)
                    # Save the above and below positions
                    if previous_position and next_position:
                        extracted_positions.append(previous_position)
                        extracted_positions.append(next_position)

    # Use the function to get the last character position
    last_position = get_last_character_position(input_file_path)

    # Add the last found position (from the text) as the ending position
    if last_position and last_position != extracted_positions[-1]:
        extracted_positions.append(last_position)

    last_line_index = len(extracted_positions) // 2  # Calculate total questions (pairs of positions)
    return extracted_positions, last_line_index

# Function to get the last character position from the file
def get_last_character_position(input_file_path):
    try:
        # Open and read the input file
        with open(input_file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print(f"Error: The file {input_file_path} was not found.")
        return None
    except Exception as e:
        print(f"An error occurred while reading the file: {e}")
        return None

    last_position = None

    # Find the last character position in the file by going through it in reverse
    for line in reversed(lines):
        match = re.match(r"(\d+):", line.strip())
        if match:
            last_position = match.group(1)
            break

    # Return the last character position
    return last_position

# Main function to process all files in the input directory and save to output directory
def process_all_files(input_folder, output_folder, total_questions_file):
    # List to hold the total questions count for each file
    total_questions_list = []
    
    # Check if the input and output directories exist
    if not os.path.exists(input_folder):
        print(f"The input folder {input_folder} does not exist.")
        return

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)  # Create the output folder if it doesn't exist

    # Loop through all files in the input folder
    for filename in os.listdir(input_folder):
        if filename.endswith('.txt'):  # Process only text files
            input_file_path = os.path.join(input_folder, filename)
            output_file_path = os.path.join(output_folder, f"{filename}")

            # Extract positions and calculate the total questions
            extracted_positions, total_questions = extract_above_below_positions(input_file_path)
            
            # Write the extracted positions to the output file
            if extracted_positions:
                try:
                    with open(output_file_path, 'w', encoding='utf-8') as output_file:
                        # Write the positions in pairs to the file
                        for i in range(0, len(extracted_positions) - 1, 2):
                            first_number = int(extracted_positions[i])
                            second_number = int(extracted_positions[i + 1]) + 1  # Add +1 to the second number
                            output_file.write(f"{first_number}, {second_number}\n")
                except Exception as e:
                    print(f"An error occurred while writing to the file {output_file_path}: {e}")
            
            # Save the total questions count for each file
            if total_questions is not None:
                total_questions_list.append(f"{filename} => Total Questions: {total_questions}")

    # Write the total questions count to "Total Questions.txt"
    with open(total_questions_file, 'w', encoding='utf-8') as total_file:
        for line in total_questions_list:
            total_file.write(line + "\n")

# Example usage
input_folder = r'C:\Users\uk880\OneDrive\Desktop\RapidPaste\5_Formatted Bullets'  # Folder containing the text files
output_folder = r'C:\Users\uk880\OneDrive\Desktop\RapidPaste\7_Question Range'  # Folder to save the extracted positions
total_questions_file = "Total Questions.txt"  # File to save the total questions

# Call the function to process all files in the folder and save the total questions count
process_all_files(input_folder, output_folder, total_questions_file)

print("Process completed and total questions saved to Total Questions.txt.")
```

```python
import os
import re

def find_sequences_from_positions(lines):
    # Initialize variables
    positions = []
    current_char_pos = None
    matched_positions = set()  # Set to track matched positions
    total_lines_used = 0  # Counter for lines contributing to output

    # Iterate through each line in the provided list of lines
    for idx, line in enumerate(lines):
        line = line.strip()  # Clean up the line
        
        # Check for opening parenthesis
        if "Character position" in line and "'('" in line:
            match = re.search(r'Character position (\d+):', line)
            if match:
                current_char_pos = int(match.group(1))  # Get the position number of '('

        # Check for characters a, b, c, d followed by a dot '.'
        elif "Character position" in line and any(char in line for char in ['a', 'b', 'c', 'd']):
            match = re.search(r'Character position (\d+): \'([abcd])\'', line)
            if match and current_char_pos is not None:
                middle_pos = int(match.group(1))  # Get the position number of 'a', 'b', 'c', or 'd'
                
                # Ensure the next line contains the dot '.'
                if idx + 1 < len(lines) and "Character position" in lines[idx + 1] and "'.'" in lines[idx + 1]:
                    end_pos = current_char_pos + 3  # Adjust for '.' and ')'
                    
                    # Ensure the next-next line contains the closing parenthesis ')'
                    if idx + 2 < len(lines) and "Character position" in lines[idx + 2] and "')'" in lines[idx + 2]:
                        # Check if the positions are not already matched
                        if (current_char_pos, middle_pos, end_pos) not in matched_positions:
                            positions.append((current_char_pos, middle_pos, end_pos))
                            matched_positions.add((current_char_pos, middle_pos, end_pos))  # Mark as matched
                            total_lines_used += 1  # Increment counter for valid output lines
                    else:
                        continue  # Skip invalid match
                else:
                    continue  # Skip invalid match

        # Check for closing parenthesis
        elif "Character position" in line and "')'" in line:
            continue  # Already handled

    return positions, total_lines_used

# Function to write only matched positions to a file
def write_positions_to_file(positions, output_filename):
    try:
        with open(output_filename, 'w', encoding='utf-8') as file:
            for start, middle, end in positions:
                file.write(f"{start}, {end + 1}\n")
        
    except Exception as e:
        print(f"An error occurred while writing to the file: {e}")

# Function to log outputs to the file Total Options.txt
def log_output(output_text, log_file_path):
    try:
        with open(log_file_path, 'a', encoding='utf-8') as log_file:
            log_file.write(output_text + "\n")
    except Exception as e:
        print(f"An error occurred while logging output: {e}")

# Main processing function for input/output directories
def process_files(input_dir, output_dir, log_file_path):
    # Ensure the output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Open the log file in 'w' mode to overwrite any existing content
    with open(log_file_path, 'w', encoding='utf-8') as log_file:
        log_file.write("")  # Clear the log file by writing an empty string

    total_questions_sum = 0  # Variable to store the sum of total questions

    # Iterate through all files in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith('.txt'):  # Process only text files
            input_filepath = os.path.join(input_dir, filename)

            # Read lines from the input file
            with open(input_filepath, 'r', encoding='utf-8', errors='replace') as file:
                lines = file.readlines()

            # Find positions and total lines used
            positions, total_lines_used = find_sequences_from_positions(lines)

            # Write results to the output file
            output_filepath = os.path.join(output_dir, filename)
            if positions:
                write_positions_to_file(positions, output_filepath)
                output_text = f"Total Options: {total_lines_used}  {filename}"
                total_questions_sum += total_lines_used * 4  # Multiply each result by 4
            else:
                output_text = f"No valid sequences found in {filename}."

            # Log the output to the Total Options.txt file
            log_output(output_text, log_file_path)

    # Log the total multiplied result
    print("Multiply above (Total Question x 4)")
    # print(f"{output_directory}")
    # log_output(f"{output_dir}", log_file_path)

# Example usage
input_directory = r'C:\Users\uk880\OneDrive\Desktop\RapidPaste\4_Clean Sequence'  # Change to your input folder path
output_directory = r'C:\Users\uk880\OneDrive\Desktop\RapidPaste\8_Options Range'  # Change to your output folder path
log_file_path = r'C:\Users\uk880\OneDrive\Desktop\RapidPaste\Total Options.txt'  # Path for the log file

process_files(input_directory, output_directory, log_file_path)

```

```python
def combine_files(file1_path, file2_path, output_file_path):
    try:
        # Open both files and read the lines
        with open(file1_path, 'r', encoding='utf-8') as file1, open(file2_path, 'r', encoding='utf-8') as file2:
            lines_file1 = file1.readlines()
            lines_file2 = file2.readlines()

        # Check that both files have the same number of lines
        if len(lines_file1) != len(lines_file2):
            print("Error: Files do not have the same number of lines.")
            return

        # Combine lines from both files
        combined_lines = []
        for line1, line2 in zip(lines_file1, lines_file2):
            combined_lines.append(line2.strip() + " -> " + line1.strip())

        # Write the combined lines to the output file
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            for combined_line in combined_lines:
                output_file.write(combined_line + "\n")

        print(f"Files successfully combined into {output_file_path}")
    
    except FileNotFoundError as e:
        print(f"Error: One of the files was not found: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
file1_path = 'Total Options.txt'  # Path to the first file
file2_path = 'Total Questions.txt'  # Path to the second file
output_file_path = 'Total_Combined.txt'  # Path to the output file

# Call the function to combine the files
combine_files(file1_path, file2_path, output_file_path)
```

```python
import os

def combine_and_sort_files(file1_path, file2_path, output_file_path):
    # Initialize a list to hold all the numbers
    combined_numbers = []

    # Read the first file
    with open(file1_path, 'r') as file1:
        for line in file1:
            # Extract numbers from the line and add to the list
            numbers = line.strip().split(', ')
            combined_numbers.extend(int(num) for num in numbers)

    # Read the second file
    with open(file2_path, 'r') as file2:
        for line in file2:
            # Extract numbers from the line and add to the list
            numbers = line.strip().split(', ')
            combined_numbers.extend(int(num) for num in numbers)

    # Sort the combined numbers
    combined_numbers.sort()

    # Write the sorted numbers to the output file
    with open(output_file_path, 'w') as output_file:
        for number in combined_numbers:
            output_file.write(f"{number}\n")

# Define the input and output folder paths
question_range_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\7_Question Range"
options_range_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\8_Options Range"
output_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\9_Order Sequence"

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all text files in the Question Range folder
for filename in os.listdir(question_range_folder):
    if filename.endswith(".txt"):
        # Define paths for both input files
        file1_path = os.path.join(question_range_folder, filename)
        file2_path = os.path.join(options_range_folder, filename)
        output_file_path = os.path.join(output_folder, filename)

        # Combine and sort the files, handling the case where the second file might not exist
        if os.path.exists(file2_path):
            combine_and_sort_files(file1_path, file2_path, output_file_path)
            print(f"Combined and sorted data has been written to {filename}")
        else:
            print(f"File {file2_path} does not exist. Skipping.")

print("All files have been processed.")
```

```python
import os

def save_number_pairs(file_path, output_file_path):
    # Initialize a list to hold all the numbers
    numbers = []

    # Read the input file
    with open(file_path, 'r') as file:
        for line in file:
            # Extract number from the line and add to the list
            number = line.strip().replace(',', '')  # Remove commas
            if number:  # Ensure there's a valid number
                numbers.append(int(number))

    # Initialize a list to hold pairs
    pairs = []

    # Create pairs from the numbers by taking two consecutive numbers at a time
    for i in range(0, len(numbers) - 1, 2):  # Step through the list two at a time
        pairs.append((numbers[i], numbers[i + 1]))

    # Write the pairs to the output file
    with open(output_file_path, 'w') as output_file:
        for pair in pairs:
            output_file.write(f"({pair[0]}, {pair[1]}),\n")

# Define the input and output folder paths
input_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\9_Order Sequence"
output_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\10_Arrangement Done"

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop through all text files in the Order Sequence folder
for filename in os.listdir(input_folder):
    if filename.endswith(".txt"):
        # Define paths for input and output files
        file_path = os.path.join(input_folder, filename)
        output_file_path = os.path.join(output_folder, filename)

        # Save number pairs from the file
        save_number_pairs(file_path, output_file_path)
        print(f"Number pairs have been written to {filename}.")

print("All files have been processed.")
```

```python
import win32com.client as win32
import os
import re
import time

def process_document(input_file_path, range_file, output_doc_template, output_folder, error_files):
    # Open MS Word Application
    word = win32.Dispatch("Word.Application")
    word.Visible = False  # Set to True if you want to see Word opening

    # Open the input document and the output template
    try:
        input_doc = word.Documents.Open(input_file_path)
        output_doc = word.Documents.Open(output_doc_template)  # Open the template (new_format1.docx)
    except Exception as e:
        print(f"Error opening documents for {input_file_path}: {e}")
        word.Quit()
        error_files.append(input_file_path)  # Log the file causing the issue
        return

    # Read the range file and parse ranges
    question_ranges = []
    tuple_pattern = re.compile(r"\((\d+),\s*(\d+)\)")

    # Check if range file exists
    if not os.path.isfile(range_file):
        print(f"Range file {range_file} not found for {input_file_path}.")
        word.Quit()
        error_files.append(input_file_path)  # Log the file causing the issue
        return

    # Open and parse the range file
    try:
        with open(range_file, 'r') as f:
            for line in f:
                match = tuple_pattern.search(line.strip())
                if match:
                    start, end = int(match.group(1)), int(match.group(2))
                    question_ranges.append((start, end))
                    # print(f"Added range for {filename}: {start}, {end}")
                else:
                    print(f"Skipping invalid line in {range_file}: {line.strip()}")
    except Exception as e:
        print(f"Error reading range file {range_file} for {input_file_path}: {e}")
        word.Quit()
        error_files.append(input_file_path)  # Log the file causing the issue
        return

    # Check if any ranges were loaded
    if not question_ranges:
        print(f"No valid ranges found in the range file for {input_file_path}.")
        word.Quit()
        error_files.append(input_file_path)  # Log the file causing the issue
        return

    # Start with the second table on the second page
    table_index = 2
    rows_in_table = 5  # Number of rows to fill before moving to the next table

    # Loop through the defined ranges and paste each range content into respective table cells
    for idx, (start_char, end_char) in enumerate(question_ranges, start=1):
        print(f"Processing range for {filename}: {start_char}, {end_char}")
        
        # Extract the text range from the input document
        text_range = input_doc.Range(Start=start_char, End=end_char)
        text_range.Copy()  # Copy the specified range
        
        # Wait briefly to ensure the clipboard is updated
        time.sleep(0.5)

        # Check if the output document has enough tables, create if necessary
        if table_index > output_doc.Tables.Count:
            output_doc.Tables.Add(output_doc.Range(0, 0), 1, 2)  # Create a new table with 1 row and 2 columns

        # Select the current table
        table = output_doc.Tables(table_index)

        # Calculate the row within the current table
        row = ((idx - 1) % rows_in_table) + 1
        col = 2  # Assuming you want to paste into column 2 for each row
        
        # Check if the current table has enough rows
        if row <= table.Rows.Count:
            try:
                table.Cell(row, col).Range.Paste()
            except Exception as e:
                print(f"Error pasting to cell at row {row}, column {col} for {input_file_path}: {e}")
                error_files.append(input_file_path)  # Log the file causing the issue
                break
        else:
            print(f"Row {row} does not exist in the table for {input_file_path}. Please add more rows if necessary.")
            error_files.append(input_file_path)  # Log the file causing the issue
            break
        
        # Move to the next table every five rows
        if idx % rows_in_table == 0 and idx != len(question_ranges):
            table_index += 1

    # Generate the output file path based on the input file name
    output_file_name = os.path.basename(input_file_path)  # Get the file name
    output_file_path = os.path.join(output_folder, output_file_name)  # Combine with output folder path

    # Save the output document with changes
    try:
        output_doc.SaveAs(output_file_path)
        print(f"Document saved as {output_file_path}.")
    except Exception as e:
        print(f"Error saving the document for {input_file_path}: {e}")
        error_files.append(input_file_path)  # Log the file causing the issue

    # Close the documents and Word application
    input_doc.Close(False)
    output_doc.Close(True)
    word.Quit()

# Define the folder paths
input_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\1_Docfile"  # Input folder containing docx files
range_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\10_Arrangement Done"  # Folder containing range files
output_folder = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\12_Fast doc"  # Folder to save the processed docx files
output_doc_template = r"C:\Users\uk880\OneDrive\Desktop\RapidPaste\Format_1000.docx"  # Path to the output template

# Ensure the output folder exists
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# List to store error files
error_files = []

# Loop through all Word files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".docx"):
        input_file_path = os.path.join(input_folder, filename)
        
        # Find the corresponding range file in the range folder
        range_file = os.path.join(range_folder, f"{os.path.splitext(filename)[0]}.txt")
        
        # Process the document with the found range file
        try:
            process_document(input_file_path, range_file, output_doc_template, output_folder, error_files)
            print(f"Processed {filename}.")
        except Exception as e:
            print(f"Failed to process {filename}: {e}")
            error_files.append(input_file_path)

# Report any files that encountered errors
if error_files:
    print("\nThe following files encountered issues during processing:")
    for error_file in error_files:
        print(error_file)
else:
    print("\nAll documents processed successfully.")
```

```python
import os
import re
import shutil

# Function to extract all numeric parts from a filename and combine them into a single number
def extract_combined_numeric_value(file_name):
    # Use regex to find all sequences of digits in the filename
    matches = re.findall(r'(\d+)', file_name)
    # If there are matches, convert to integer after combining them as a single number
    if matches:
        combined_number = int(''.join(matches))
        return combined_number
    else:
        return float('inf')  # If no number is found, return a high value for sorting at the end

# Function to rename files based on their sorted numerical order
def rename_and_copy_files(input_folder, starting_number, output_folder):
    # Ensure the output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get all files in the folder and filter out non-files
    files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]

    # Sort the files using the combined numeric value extracted from the filenames
    files_sorted = sorted(files, key=extract_combined_numeric_value)

    # Log the order of files after sorting for verification
    print("\nOrder of files after sorting based on the combined numeric value:")
    for file in files_sorted:
        print(file)

    # Rename and copy each file to the output directory
    for i, file_name in enumerate(files_sorted):
        # Get the file extension
        file_extension = os.path.splitext(file_name)[1]

        # Create the new file name with an incrementing index starting from the provided number
        new_file_name = f"CH-{starting_number + i}{file_extension}"

        # Full paths for the input and output
        old_file_path = os.path.join(input_folder, file_name)
        new_file_path = os.path.join(output_folder, new_file_name)

        # Copy and rename the file
        shutil.copy2(old_file_path, new_file_path)
        print(f"Copied and renamed: {file_name} -> {new_file_name}")

# Example usage
input_folder = "12_Fast doc"  # Folder containing the original files
output_folder = "12_FileName Update"       # Folder where renamed files will be saved
starting_number = 1                     # Starting number for the new names

# Execute the function
rename_and_copy_files(input_folder, starting_number, output_folder)
```
